```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
z1 = stats.norm.ppf(q=0.975, loc=0, scale=1)
interval1 = (4.2-z1*np.sqrt(1.44)/8,4.2+z1*np.sqrt(1.44)/8)
print(interval1)

z2 = stats.norm.ppf(q=0.995, loc=0, scale=1)
interval2 = (4.2-z2*np.sqrt(1.44)/8,4.2+z2*np.sqrt(1.44)/8)
print(interval2)

print(np.diff(interval1))
print(np.diff(interval2))
```

### testowanie hipotez

```python
alpha = 0.05
X_mean = 1.54
mu0 = 1.6
sigma = 0.8
n = 20
```

```python
T = (X_mean-mu0)/(sigma/np.sqrt(n))
print(T)
```

```python
z = stats.norm.ppf(q=1-alpha, loc=0, scale=1)
print(-z)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 10))

norm01 = stats.norm(loc=0, scale=1)

x0 = -z
a, b = -5, 5
x_value = np.linspace(a, b, 100)

x = np.linspace(a, x0)
y = norm01.pdf(x)

ax.plot(x_value, norm01.pdf(x_value))
ax.fill_between(x, y, alpha=0.2)
ax.text(x0, 0.05, "z={}".format(np.round(x0, 2)), size=30)
ax.vlines(T, 0, norm01.pdf(T), linestyles='--')
ax.text(T, 0.1, "T={}".format(np.round(T, 2)), size=30)
ax.set_xlabel('x')
ax.set_ylabel('PDF(x)')
plt.show()
```
