```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
from matplotlib import pyplot as plt
```

```python
np.random.seed(42)
data = np.random.randn(100).cumsum()
ts = pd.Series(data, index=pd.date_range(start='2020-01-01', periods=len(data), freq='D'))
print(ts.head())
```

```python
td = pd.DataFrame(ts)
td
```

```python
plt.plot(td.index, td[0])
```

```python
result = adfuller(td)

print('Statystyka testowa: %f' % result[0])
print('p-wartosc: %f' %result[1])
print('Liczba opoznien uzyta: %d' %result[2])
print('Liczba obserwacji: %d' %result[3])
print('wartosc krytyczne:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))

if result[0] < result[4]['5%']:
    print('Szereg czasowy jest stacjonarny')
else:
    print('Szereg czasowy nie jest stacjonarny')
```
