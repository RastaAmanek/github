```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
```

```python
X = [-3, -1, 3, 5]
p = [0.1, 0.2, 0.5, 0.2]
plt.scatter(x=X, y=p)
plt.vlines(X, 0, p, colors='g')
```

```python
F = np.cumsum(p)

x_lim = [-5] + X + [10]
f = [0] + list(F)

plt.scatter(X, F, label='cdf', color='r')
plt.scatter(X, f[:-1], s=80, facecolors='none', edgecolors='r')
plt.xlim([-5, 10])
plt.ylim([-0.1, 1.1])

for i in range(len(x_lim) - 1):
    plt.hlines(xmin=x_lim[i], xmax=x_lim[i+1], y=f[i], color='black', alpha=0.5, linestyle='--')

plt.show()
```

```python
X = np.array(X)
p = np.array(p)

ex = 0.1 * -3 + 0.2 * -1 + 0.5 * 3 + 0.2 * 5
ex_ = np.sum(X*p)
print("Wartosc oczekiwana: {}".format(ex))
print("lub: {}".format(ex_))
```

```python
ex2 = np.sum((X**2)*p)
print("Drugi moment: {}".format(ex2))
```

```python
var = ex2 - ex**2
print("Wariancja: {}".format(var))
```

```python
std = np.sqrt(var)
print("Odchylenie standardowe: {}".format(std))
```

```python
fig, ax = plt.subplots(1,1)

n, p = 6, 0.5

binom = stats.binom(n=n, p=p)

ax.set_xlim([-10, 10])
# ay.set_ylim([-0.2, 1.2])

x = np.arange(-10, 10, 0.01)
ax.plot(x, binom.cdf(x), 'go--', ms=3, label='bernoulli cdf')
plt.show()
```

```python
N = 1000
n_vec = np.arange(5, 1000, 1)
var0, var1 = np.zeros(len(n_vec)), np.zeros(len(n_vec))

for i, n in enumerate(n_vec):
    var0_vec, var1_vec = np.zeros(N), np.zeros(N)
    for j in np.arange(N):
        x = stats.norm.rvs(size=n, loc=-4, scale=3)
        var0_vec[j] = np.var(x, ddof = 0)
        var1_vec[j] = np.var(x, ddof = 1)
    var0[i] = np.mean(var0_vec)
    var1[i] = np.mean(var1_vec)
```

```python
plt.plot(var0, label = '$S^2_n$')
plt.plot(var1, label = '$S^2$')
plt.xticks(ticks=np.arange(0, len(n_vec)+5, 5), labels=np.arange(5, 1005, 5))
plt.hlines(xmin=0, xmax=1000, y=9, linestyles='--')
plt.legend()
plt.show()
```
